```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_excel(r"E:\ALEX\Customer_Call_List.xlsx")
df
```

```python
df.drop("Not_Useful_Column",axis=1).head()
```

```python
df.head()
```

```python
df.drop(columns="Not_Useful_Column",inplace=True)
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df["Last_Name"]=df["Last_Name"].str.strip(";,/...")
```

```python
df.head()
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace("[^A-Za-z0-9]","",regex=True)
```

```python
df.head()
```

```python
type(df["Phone_Number"][2])
```

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: str(x))
```

```python
type(df["Phone_Number"][2])
```

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: x[0:3]+"-"+x[3:6]+"-"+x[6:])
df.head()
```

```python
df["Phone_Number"]=df["Phone_Number"].replace("nan--","")
df["Phone_Number"]=df["Phone_Number"].replace("Na--","")
```

```python
df[["Street","State","Zip_Code"]]=df["Address"].str.split(",",n=2,expand=True)
df.head()
```

```python
df=df.drop(columns="Address")
df.head()
```

```python
df["Paying Customer"]=df["Paying Customer"].replace("N","No")
df["Paying Customer"]=df["Paying Customer"].replace("Y","Yes")
df.head()
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].replace("N","No")
df["Do_Not_Contact"]=df["Do_Not_Contact"].replace("Y","Yes")
df.head()
```

```python
for x in df.index:
    if df.loc[x,"Do_Not_Contact"]=="Yes":
        df.drop(x,inplace=True)
df
```

```python
import numpy as np
df["Do_Not_Contact"]=df["Do_Not_Contact"].replace(np.nan,"")
```

```python
for x in df.index:
    if df.loc[x,"Do_Not_Contact"]=="":
        df.drop(x,inplace=True)
df
```

```python
df.sort_index(inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x,"Phone_Number"]=="":
        df.drop(x,inplace=True)
df
```

```python
df.reset_index(drop=True,inplace=True)
```
